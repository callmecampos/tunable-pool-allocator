# Quantifying Runtime Performance for Different Optimizations

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

### Last Used Pool Caching (ran on macOS Catalina 10.15.4)

```python
num_pools, runs_per_call = 64, 1000
binary_cache = np.array([1.95, 1.72, 1.72, 1.71, 1.73, 1.69, 1.69, 1.69, 1.74, 1.72])
binary_search = np.array([1.90, 1.77, 1.80, 1.82, 1.82, 1.78, 1.81, 1.82, 1.77, 1.80])
naive_search = np.array([1.99, 1.85, 1.85, 1.84, 1.84, 1.86, 1.82, 1.84, 1.87, 1.88])
len(binary_cache), len(binary_search), len(naive_search)
```

```python
plt.plot(naive_search, ":", color="red", label="Linear Search")
plt.plot(binary_search, "--", color="C0", label="Binary Search")
plt.plot(binary_cache, color="C0", label="Binary Search with Cache")
plt.xlabel(f"Test # ({runs_per_call} runs per test)")
plt.ylabel("Runtime (s)")
plt.title(f"pool_alloc() Runtime")
plt.legend()
plt.savefig("pool_alloc_runtime.png")
```

### Lazy Initialization (ran on macOS Catalina 10.15.4)

```python
num_pools, runs_per_call = 64, 2000
lazy_init = np.array([1.60, 1.44, 1.41, 1.43, 1.42, 1.46, 1.45, 1.43, 1.43, 1.44])
regular_init = np.array([1.67, 1.53, 1.53, 1.53, 1.52, 1.52, 1.53, 1.55, 1.52, 1.52])
len(lazy_init), len(regular_init)
```

```python
plt.plot(regular_init, "--", color="C0", label="Regular Init")
plt.plot(lazy_init, color="C0", label="Lazy Init")
plt.xlabel(f"Test # ({runs_per_call} runs per test)")
plt.ylabel("Runtime (s)")
plt.title(f"pool_init() Runtime")
plt.legend()
plt.savefig("pool_init_runtime.png")
```
